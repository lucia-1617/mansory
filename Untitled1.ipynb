**Escaneo de archivos con downsampling de “sin_grietas”**

```python
# Celda 3
# Celda 3 — ESCANEO con mapeo exacto de carpetas

# Nombres canónicos de clases (10) que usaremos en el modelo
CLASS_NAMES = [
    "compresion_vertical_Grave",
    "compresion_vertical_Leve",
    "compresion_vertical_Moderada",
    "friccion_cortante_escalonada_Grave",
    "friccion_cortante_escalonada_Leve",
    "friccion_cortante_escalonada_Moderada",
    "sin_grietas",
    "tension_diagonal_inclinadas_Grave",
    "tension_diagonal_inclinadas_Leve",
    "tension_diagonal_inclinadas_Moderada",
]

# Mapeo de prefijo de clase → nombre EXACTO de la carpeta en disco
FOLDER_MAP = {
    "compresion_vertical":           "compresion_vertical",
    "friccion_cortante_escalonada":  "friccion cortante_escalonada",  # ojo: espacio + _
    "sin_grietas":                   "sin grietas",                    # espacio
    "tension_diagonal_inclinadas":   "tension diagonal_inclinadas",   # espacio + _
}

# Límite por clase en TRAIN (downsample fuerte de 'sin_grietas' para equilibrar)
CAPS_TRAIN = {"sin_grietas": 3000}
CAPS_VAL   = {}   # sin límite

EXTS = (".jpg",".jpeg",".png",".bmp",".JPG",".JPEG",".PNG",".BMP")

def split_prefix_sev(class_name: str):
    """'compresion_vertical_Grave' -> ('compresion_vertical','Grave')"""
    if class_name == "sin_grietas":
        return "sin_grietas", None
    pref, sev = class_name.rsplit("_", 1)
    return pref, sev

def scan_with_map(root: Path, caps: dict):
    files, labels = [], []
    counts = []

    for idx, cname in enumerate(CLASS_NAMES):
        pref, sev = split_prefix_sev(cname)
        folder = FOLDER_MAP[pref]  # nombre real en disco

        if cname == "sin_grietas":
            d = root / folder
            imgs = []
            for e in EXTS: imgs += list(d.glob(f"*{e}"))
        else:
            d = root / folder / sev
            imgs = []
            for e in EXTS: imgs += list(d.glob(f"*{e}"))

        # control de existencia para depurar
        if not d.exists():
            print(f"⚠️  Carpeta NO encontrada: {d}")

        # límite por clase si aplica
        cap = caps.get(cname, None) if caps else None
        if cap and len(imgs) > cap:
            imgs = imgs[:cap]

        files.extend(imgs)
        labels.extend([idx] * len(imgs))
        counts.append(len(imgs))

    return files, labels, np.array(counts, dtype=np.int64)

train_files, train_labels, train_counts = scan_with_map(TRAIN_DIR, CAPS_TRAIN)
val_files,   val_labels,   val_counts   = scan_with_map(VAL_DIR,   CAPS_VAL)

print("Clases (orden):")
for i, n in enumerate(CLASS_NAMES):
    print(f"{i:2d} -> {n}")

print("\nTrain por clase:", train_counts.tolist())
print("Val   por clase:", val_counts.tolist())

```

**Dataset rápido (cache, prefetch) y preprocesado**

```python
# Celda 4
def load_and_preprocess(path, label, training: bool):
    img = tf.io.read_file(path)
    img = tf.io.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.convert_image_dtype(img, tf.float32)      # [0,1]
    img = tf.image.resize(img, IMG_SIZE)
    if training:
        img = tf.image.random_flip_left_right(img)
        img = tf.image.random_brightness(img, 0.08)
        img = tf.image.random_contrast(img, 0.9, 1.1)
    # Escala [-1,1] (MobileNetV2)
    img = (img - 0.5) * 2.0
    return img, label

def make_ds(files, labels, training=True):
    files = list(map(str, files))
    ds = tf.data.Dataset.from_tensor_slices((files, labels))
    if training:
        ds = ds.shuffle(min(len(files), 10000), reshuffle_each_iteration=True)
    ds = ds.map(lambda p,l: load_and_preprocess(p,l,training),
                num_parallel_calls=AUTOTUNE)
    ds = ds.batch(BATCH)
    ds = ds.prefetch(AUTOTUNE)
    return ds

train_ds = make_ds(train_files, train_labels, training=True)
val_ds   = make_ds(val_files,   val_labels,   training=False)
```

**Modelo (MobileNetV2 liviano) + Focal Loss con α**

```python
# Celda 5
num_classes = len(CLASS_NAMES)

# alpha (inversa de frecuencia, normalizada y acotada)
counts = train_counts + 1e-9
alpha  = (counts.sum() / (num_classes * counts))
alpha  = alpha / alpha.mean()
alpha  = np.clip(alpha, 0.25, 4.0).astype("float32")
print("alpha:", np.round(alpha, 3).tolist())

def focal_loss(alpha_vec, gamma=2.0):
    alpha_const = tf.constant(alpha_vec, dtype=tf.float32)
    def loss(y_true, y_pred):
        y_true = tf.cast(y_true, tf.int32)
        y_true = tf.one_hot(y_true, depth=num_classes)
        y_pred = tf.clip_by_value(y_pred, 1e-7, 1.0-1e-7)
        ce  = -tf.reduce_sum(y_true * tf.math.log(y_pred), axis=1)
        pt  = tf.reduce_sum(y_true * y_pred, axis=1)
        at  = tf.reduce_sum(y_true * alpha_const, axis=1)
        fl  = at * tf.pow(1.0 - pt, gamma) * ce
        return tf.reduce_mean(fl)
    return loss

base = keras.applications.MobileNetV2(
    input_shape=IMG_SIZE+(3,), include_top=False, weights="imagenet", alpha=0.35
)
base.trainable = False

inputs  = keras.Input(shape=IMG_SIZE+(3,), name="image")
x       = base(inputs, training=False)
x       = keras.layers.GlobalAveragePooling2D()(x)
x       = keras.layers.Dropout(0.25)(x)
outputs = keras.layers.Dense(num_classes, activation="softmax")(x)
model   = keras.Model(inputs, outputs, name="mansory_mobilenetv2")

model.compile(
    optimizer=keras.optimizers.Adam(1e-3),
    loss=focal_loss(alpha, gamma=2.0),
    metrics=["accuracy"]
)
model.summary()
```

**Callbacks y fit (épocas cortas en CPU)**

```python
# Celda 6
class SaveBest95(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        val_acc = (logs or {}).get("val_accuracy")
        if val_acc is not None and val_acc >= 0.95:
            self.model.save(MODEL_BEST95)
            print(f"\n✅ Guardado >=95%: {MODEL_BEST95}")

cbs = [
    keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True, monitor="val_accuracy"),
    keras.callbacks.ReduceLROnPlateau(patience=2, factor=0.5, monitor="val_accuracy"),
    keras.callbacks.ModelCheckpoint(MODEL_REGULAR, save_best_only=True, monitor="val_accuracy"),
    SaveBest95(),
]

STEPS_PER_EPOCH = max(1, EFFECTIVE_IMAGES_PER_EPOCH // BATCH)
print("STEPS_PER_EPOCH:", STEPS_PER_EPOCH)

history = model.fit(
    train_ds.repeat(),                 # permite steps_per_epoch
    steps_per_epoch=STEPS_PER_EPOCH,
    validation_data=val_ds,
    epochs=30,
    callbacks=cbs,
    verbose=1
)

# guardar nombres de clases
with open(CLASSES_JSON, "w", encoding="utf-8") as f:
    json.dump({"class_names": CLASS_NAMES}, f, ensure_ascii=False, indent=2)

print("\nClases guardadas en", CLASSES_JSON)
best_val = max(history.history.get("val_accuracy", [0.0]))
print(f"Mejor val_accuracy: {best_val:.3f}")
print("Modelo mejor por val_acc:", MODEL_REGULAR)
if MODEL_BEST95.exists():
    print("También se guardó ≥95%:", MODEL_BEST95)
```

**Inferencia (tipo + severidad SIN porcentajes)**

```python
# === Verificar y elegir el mejor modelo (y función de inferencia) ===
from pathlib import Path
import json, numpy as np
import tensorflow as tf
from tensorflow import keras

# Rutas
PROJECT_ROOT = Path.cwd()
MODEL_DIR    = PROJECT_ROOT / "modelos"
MODEL_REG    = MODEL_DIR / "modelo_mansory.keras"
MODEL_95     = MODEL_DIR / "modelo_mansory_95.keras"
CLASSES_JSON = MODEL_DIR / "classes.json"

TRAIN_DIR = PROJECT_ROOT / "train"
VAL_DIR   = PROJECT_ROOT / "test"   # tu validación está en 'test'

# Config (mismo preprocesado que entrenamiento)
IMG_SIZE = (160, 160)
def preprocess(x):
    x = tf.image.resize(x, IMG_SIZE)
    x = tf.image.convert_image_dtype(x, tf.float32)
    return (x - 0.5) * 2.0

# --- nombres de clases (orden) ---
class_names = json.loads(CLASSES_JSON.read_text(encoding="utf-8"))["class_names"]
num_classes = len(class_names)
print("Clases (orden):", class_names)

# --- mapeo EXACTO de carpetas ---
FOLDER_MAP = {
    "compresion_vertical":           "compresion_vertical",
    "friccion_cortante_escalonada":  "friccion cortante_escalonada",
    "sin_grietas":                   "sin grietas",
    "tension_diagonal_inclinadas":   "tension diagonal_inclinadas",
}
EXTS = (".jpg",".jpeg",".png",".bmp",".JPG",".JPEG",".PNG",".BMP")

def split_prefix_sev(cn:str):
    if cn=="sin_grietas": return "sin_grietas", None
    pref, sev = cn.rsplit("_", 1)
    return pref, sev

def scan_with_map(root: Path):
    files, labels = [], []
    for idx, cname in enumerate(class_names):
        pref, sev = split_prefix_sev(cname)
        base = root / FOLDER_MAP[pref]
        d = base if cname=="sin_grietas" else base / sev
        imgs=[]
        for e in EXTS: imgs += list(d.glob(f"*{e}"))
        files += imgs
        labels += [idx]*len(imgs)
    return files, labels

val_files, val_labels = scan_with_map(VAL_DIR)
print(f"Validación: {len(val_files)} imágenes")

# TF Dataset
def load_img_from_path(p):
    x = tf.io.read_file(p)
    x = tf.io.decode_image(x, channels=3, expand_animations=False)
    return preprocess(x)

def make_ds(files, labels, batch=64):
    pds = tf.data.Dataset.from_tensor_slices([str(p) for p in files])
    xds = pds.map(load_img_from_path, num_parallel_calls=1)  # 1 hilo: más estable en CPU
    yds = tf.data.Dataset.from_tensor_slices(labels)
    ds  = tf.data.Dataset.zip((xds, yds)).batch(batch).prefetch(tf.data.AUTOTUNE)
    return ds

val_ds = make_ds(val_files, val_labels, batch=64)

def eval_model(path: Path, name: str):
    if not path.exists():
        print(f"{name}: no encontrado -> {path}")
        return None, None
    try:
        # Cargar SIN compilar (evita el error "Could not locate function 'loss'")
        m = keras.models.load_model(str(path), compile=False, safe_mode=False)
    except Exception as e:
        print(f"Falló load_model({path}): {e}")
        return None, None

    # Recompilar con la misma configuración que usamos para entrenar
    # (salida softmax + etiquetas enteras => 'sparse_categorical_crossentropy')
    m.compile(
        optimizer="adam",
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )

    loss, acc = m.evaluate(val_ds, verbose=0)
    print(f"{name:>22s}: val_acc={acc:.4f}  (loss={loss:.4f})  -> {path}")
    return acc, m


print("\nEvaluando modelos guardados:")
acc_reg, m_reg = eval_model(MODEL_REG, "modelo_mansory.keras")
acc_95,  m_95  = eval_model(MODEL_95,  "modelo_mansory_95.keras")

# Elegir el mejor real
best_model = None
best_name  = ""
if acc_reg is not None and (acc_95 is None or acc_reg >= acc_95):
    best_model, best_name = m_reg, "modelo_mansory.keras"
if acc_95 is not None and (acc_reg is None or acc_95 > acc_reg):
    best_model, best_name = m_95,  "modelo_mansory_95.keras"

if best_model is None:
    raise RuntimeError("No se encontró ningún modelo .keras en la carpeta 'modelos'.")

print(f"\n✅ Usaremos: {best_name}")

# ---- Inferencia en una imagen suelta (sin porcentajes) ----
def split_type_severity(s: str):
    s = s.strip().replace("__","_")
    if s == "sin_grietas": return "sin grietas", "—"
    parts = s.split("_")
    return " ".join(parts[:-1]).replace("  "," "), parts[-1]

def predict_image(img_path):
    from tensorflow.keras.utils import load_img, img_to_array
    img = load_img(img_path, target_size=IMG_SIZE)
    x   = img_to_array(img)
    x   = preprocess(tf.convert_to_tensor(x)).numpy()
    x   = np.expand_dims(x, 0)

    pred = best_model.predict(x, verbose=0)
    pred_id  = int(np.argmax(pred[0]))
    nombre   = class_names[pred_id]
    tipo, sev = split_type_severity(nombre)

    import matplotlib.pyplot as plt
    plt.figure(figsize=(4,5))
    plt.imshow(load_img(img_path))
    plt.axis("off")
    plt.title(f"Tipo de grieta: {tipo}\nSeveridad: {sev}", fontsize=12)
    plt.show()
    print(f"Archivo: {img_path}\nTipo: {tipo}\nSeveridad: {sev}")

# Ejemplo:
# predict_image(PROJECT_ROOT / "55.jpg")
```

**Inferencia (muestra solo “Tipo de grieta” y “Severidad”)**

```python
# --- Inferencia con tu modelo 10 clases + multi-crops + umbral contra "sin_grietas" ---

from pathlib import Path
import json, numpy as np
from PIL import Image, ImageOps
import keras
from keras.utils import img_to_array

MODEL_DIR = Path("modelos")
# elegimos el mejor disponible
candidatos = [
    MODEL_DIR / "modelo_mansory_95.keras",
    MODEL_DIR / "modelo_mansory.keras",
    Path("modelo_mansory_95.keras"),
    Path("modelo_mansory.keras"),
]
MODEL_PATH = next((p for p in candidatos if p.exists()), None)
assert MODEL_PATH is not None, "No encontré modelo .keras en modelos/"

# cargar modelo
model = keras.models.load_model(str(MODEL_PATH), compile=False)
H, W = model.input_shape[1], model.input_shape[2]   # tamaño correcto del modelo

# cargar nombres de clase
classes_jsons = [MODEL_DIR/"classes.json", Path("classes.json"), Path("clases_mansory.json")]
class_names = None
for cj in classes_jsons:
    if cj.exists():
        class_names = json.loads(cj.read_text(encoding="utf-8"))["class_names"]
        break
assert class_names is not None, "No encontré classes.json/clases_mansory.json con 'class_names'."

# índice de "sin_grietas"
try:
    IDX_SIN = class_names.index("sin_grietas")
except ValueError:
    raise RuntimeError(f"'sin_grietas' no está en class_names: {class_names}")

THRESHOLD_CRACK = 0.55  # baja este umbral (0.50/0.45) si aún te dice 'sin grietas' en grietas finas

def _preprocess(arr):
    x = arr.astype("float32") / 255.0
    x = (x - 0.5) * 2.0
    return np.expand_dims(x, 0)

def _five_crops_pil(pil_img, crop_ratio=0.8):
    w, h = pil_img.size
    cw, ch = int(w*crop_ratio), int(h*crop_ratio)
    boxes = [
        (0, 0, cw, ch),                 # top-left
        (w-cw, 0, w, ch),               # top-right
        (0, h-ch, cw, h),               # bottom-left
        (w-cw, h-ch, w, h),             # bottom-right
        ((w-cw)//2, (h-ch)//2, (w+cw)//2, (h+ch)//2),  # center
    ]
    crops = [pil_img.crop(b).resize((W, H)) for b in boxes]
    # añadir flips horizontales
    crops += [ImageOps.mirror(c) for c in crops]
    return crops

def predict_image(path_img):
    pil = Image.open(path_img).convert("RGB")
    crops = _five_crops_pil(pil, crop_ratio=0.80)

    # promediar probabilidades sobre los 10 crops
    probs = []
    for c in crops:
        x = _preprocess(img_to_array(c))
        p = model.predict(x, verbose=0)[0]      # softmax 10 clases
        probs.append(p)
    p_mean = np.mean(probs, axis=0)

    p_sin = float(p_mean[IDX_SIN])
    p_crack = 1.0 - p_sin

    if p_crack < THRESHOLD_CRACK:
        print(f"{path_img}\n→ Tipo: sin grietas | Severidad: —  (p_con_grieta={p_crack:.3f})")
        return "sin grietas", "—", p_crack

    # elegimos la mejor clase entre TODAS menos sin_grietas
    p_mean_masked = p_mean.copy()
    p_mean_masked[IDX_SIN] = -1.0
    idx = int(np.argmax(p_mean_masked))
    nombre = class_names[idx]

    # separar tipo / severidad (último token es la severidad)
    parts = nombre.split("_")
    tipo = " ".join(parts[:-1]).replace("_"," ")
    sev  = parts[-1].replace("_"," ")
    print(f"{path_img}\n→ Tipo: {tipo} | Severidad: {sev}  (conf={p_mean[idx]:.3f}, p_con_grieta={p_crack:.3f})")
    return tipo, sev, float(p_mean[idx])

# ---- EJEMPLO ----
# Cambia la ruta a tu imagen
# predict_image("grieta-muro.jpg")


```

**Cargar modelo y utilidades (corrige el texto “sin grietas”)**

```python
from pathlib import Path
import json, numpy as np
from tensorflow import keras
from tensorflow.keras.utils import load_img, img_to_array
import matplotlib.pyplot as plt

# --- Rutas ---
ROOT = Path(r"C:\Users\User\mansory")
MODEL_DIR = ROOT / "modelos"

# Escoge automáticamente el mejor disponible
MODEL_PATH = (MODEL_DIR / "modelo_mansory_95.keras") if (MODEL_DIR / "modelo_mansory_95.keras").exists() else (MODEL_DIR / "modelo_mansory.keras")
CLASSES_JSON = MODEL_DIR / "classes.json"

# Cargar clases
class_names = json.loads(CLASSES_JSON.read_text(encoding="utf-8"))["class_names"]

# Cargar modelo SOLO para inferencia
model = keras.models.load_model(str(MODEL_PATH), compile=False)
H, W = model.input_shape[1], model.input_shape[2]
assert model.output_shape[-1] == len(class_names), "El número de salidas del modelo no coincide con classes.json"

def split_type_severity(name: str):
    """Formatea salida humana y corrige 'sin_grietas'."""
    if name == "sin_grietas":
        return "sin grietas", "—"
    s = name.replace("__", "_")
    p = s.split("_")
    # casos compuestos
    if p[0] == "friccion" and len(p) >= 4:
        tipo = "_".join(p[:3])
        sev  = p[3]
    elif p[0] == "tension" and len(p) >= 4:
        tipo = "_".join(p[:3])
        sev  = p[3]
    else:
        tipo = "_".join(p[:-1])
        sev  = p[-1]
    return tipo.replace("_", " "), sev

def preprocess(path):
    img = load_img(path, target_size=(H, W))
    x = img_to_array(img)/255.0
    x = (x - 0.5) * 2.0
    return np.expand_dims(x, 0), img
```

**Predicción con top-k (y texto arreglado)**

```python
from numpy import argsort

IMG = r"C:\Users\User\mansory\grieta-muro.jpg"  # <-- cambia aquí si quieres

x, PILimg = preprocess(IMG)
probs = model.predict(x, verbose=0)[0]
topk = 5
idxs = argsort(probs)[-topk:][::-1]

pred_id = int(idxs[0])
pred_name = class_names[pred_id]
tipo, sev = split_type_severity(pred_name)

plt.figure(figsize=(5,6))
plt.imshow(PILimg); plt.axis("off")
plt.title(f"Tipo: {tipo}\nSeveridad: {sev}", fontsize=12)
plt.show()

print("Top-5:")
for i in idxs:
    n = class_names[i]
    t, s = split_type_severity(n)
    print(f"  {t:35s} {s:10s} -> {probs[i]*100:6.2f}%")

# Umbral de prudencia para 'sin grietas'
if pred_name == "sin_grietas" and probs[pred_id] < 0.98:
    print("\n⚠️ Aviso: el modelo no está muy seguro siendo 'sin grietas'. Revisa visualmente o considera el flujo binario.")
```

```python
# === Celda extra: inferencia robusta con mosaicos + TTA y umbral ===
# === Inferencia robusta con PIL (sin OpenCV) ===
from PIL import Image
import numpy as np

def _preprocess_rgb(arr, target_w, target_h):
    im = Image.fromarray(arr)
    im = im.resize((target_w, target_h), Image.BILINEAR)
    x = np.asarray(im).astype(np.float32) / 255.0
    x = (x - 0.5) * 2.0
    return np.expand_dims(x, 0)

def _predict_patch_rgb_pil(patch_rgb, tta=True):
    x = _preprocess_rgb(patch_rgb, W, H)
    if tta:
        x_flip = x[:, :, ::-1, :]
        p1 = model.predict(x,       verbose=0)[0]
        p2 = model.predict(x_flip,  verbose=0)[0]
        p  = (p1 + p2) / 2.0
    else:
        p  = model.predict(x, verbose=0)[0]
    pid = int(np.argmax(p))
    return class_names[pid], float(p[pid]), p

def infer_with_patches_pil(img_path, crack_threshold=0.60, tta=True):
    im = Image.open(img_path).convert("RGB")
    Wimg, Himg = im.size

    # 1) Pred global
    name_g, p_g, _ = _predict_patch_rgb_pil(np.asarray(im), tta=tta)
    if name_g != "sin_grietas" and p_g >= crack_threshold:
        return name_g, p_g, None

    # 2) Escaneo por parches
    scales = [1.0, 0.8, 0.6]
    stride = max(32, int(min(H, W) * 0.5))  # 50% de solape
    best = ("sin_grietas", 0.0, None)

    for sc in scales:
        w = int(Wimg * sc); h = int(Himg * sc)
        ims = im.resize((w, h), Image.BILINEAR)
        for y in range(0, max(1, h - H + 1), stride):
            for x in range(0, max(1, w - W + 1), stride):
                patch = ims.crop((x, y, x + W, y + H))
                name, p, _ = _predict_patch_rgb_pil(np.asarray(patch), tta=tta)
                if name != "sin_grietas" and p > best[1]:
                    best = (name, p, (x, y, sc))

    # 3) Decisión
    if best[0] != "sin_grietas":
        return best
    else:
        return name_g, p_g, None

# --- Ejecutar ---
IMG = r"C:\Users\User\mansory\grieta-muro.jpg"  # cambia si quieres
name, prob, loc = infer_with_patches_pil(IMG, crack_threshold=0.60, tta=True)
tipo, sev = split_type_severity(name)
print(f"Tipo de grieta: {tipo}")
print(f"Severidad: {sev}")
print(f"Confianza: {prob*100:.2f}%")
if loc:
    x, y, sc = loc
    print(f"(Detectada en parche: x={x}, y={y}, escala={sc})")
elif name == "sin_grietas":
    print("⚠️ Sigue saliendo 'sin grietas'. Si visualmente hay fisura, reentrenar con dataset balanceado/two-stage ayudará.")

```
