**Escaneo del dataset + caps por clase (recorta â€œsin_grietasâ€)**

```python
# ==== Celda 3 (reemplazo): escaneo robusto con soporte 1 o 2 niveles ====
from pathlib import Path
from collections import defaultdict, Counter
import glob, re, json

IMG_EXTS = [".jpg", ".jpeg", ".png", ".bmp", ".webp"]

def _norm(s: str) -> str:
    # normaliza a underscores y sin dobles espacios
    s = s.strip().replace("-", "_").replace(" ", "_")
    s = re.sub("_+", "_", s)
    return s

SEV_MAP = {
    "grave": "Grave",
    "leve": "Leve",
    "moderada": "Moderada",
    "moderado": "Moderada",   # por si acaso
}

def _sev_std(name: str) -> str | None:
    k = _norm(name).lower()
    k = k.replace("_", "")
    for key, val in SEV_MAP.items():
        if k == key:
            return val
    return None

def scan_dataset_two_levels(root: Path) -> dict[str, list[str]]:
    """
    Lee dataset en dos formatos:
    A) Plano:   root/<clase>/*.jpg  (p. ej., 'friccion_cortante_escalonada_Grave')
    B) Anidado: root/<tipo>/<sev>/*.jpg (p.ej. 'friccion_cortante_escalonada/Grave/*.jpg')
    Devuelve: dict{ class_name -> [rutas] }
    """
    root = Path(root)
    out = defaultdict(list)
    # 1) detecta si hay nivel de severidades
    tipo_dirs = [d for d in root.iterdir() if d.is_dir()]
    has_nested = False
    for d in tipo_dirs:
        subdirs = [sd for sd in d.iterdir() if sd.is_dir()]
        if any(_sev_std(sd.name) for sd in subdirs):
            has_nested = True
            break

    if not has_nested:
        # A) plano
        for d in tipo_dirs:
            cname = _norm(d.name)
            for ext in IMG_EXTS:
                out[cname].extend(glob.glob(str(d / f"*{ext}")))
    else:
        # B) anidado: tipo/sev/*.*
        for tipo in tipo_dirs:
            tipo_norm = _norm(tipo.name)
            if tipo_norm.lower() in {"sin_grietas", "sin_grieta", "no_cracks"}:
                # 'sin_grietas' puede no tener subcarpetas
                for ext in IMG_EXTS:
                    out["sin_grietas"].extend(glob.glob(str(tipo / f"*{ext}")))
                # y, si tuviera por error subcarpetas, tambiÃ©n las recogemos
                for sev in [sd for sd in tipo.iterdir() if sd.is_dir()]:
                    for ext in IMG_EXTS:
                        out["sin_grietas"].extend(glob.glob(str(sev / f"*{ext}")))
                continue

            for sev in [sd for sd in tipo.iterdir() if sd.is_dir()]:
                sev_std = _sev_std(sev.name)
                if not sev_std:
                    # si no es Grave/Leve/Moderada, lo ignoramos
                    continue
                class_name = f"{tipo_norm}_{sev_std}"
                for ext in IMG_EXTS:
                    out[class_name].extend(glob.glob(str(sev / f"*{ext}")))
    return out

def summarize_split(name: str, root: Path):
    data = scan_dataset_two_levels(root)
    # Orden estable por nombre
    classes = sorted(data.keys())
    print(f"{name} â€” clases detectadas ({len(classes)}): {classes}")
    cnt = {c: len(data[c]) for c in classes}
    print(f"{name} â€” total imÃ¡genes: {sum(cnt.values())}")
    print(f"{name} â€” imÃ¡genes por clase:")
    for c in classes:
        print(f"  {c:35s} -> {cnt[c]}")
    return classes, cnt, data

# --- Ejecutar resumen para TRAIN y VAL (o TEST) ---
print("RaÃ­z TRAIN:", TRAIN_DIR)
print("RaÃ­z VAL  :", VAL_DIR)

train_classes, train_counts, train_data = summarize_split("TRAIN", TRAIN_DIR)
val_classes,   val_counts,   val_data   = summarize_split("VAL",   VAL_DIR)

# ComprobaciÃ³n mÃ­nima esperada:
# - Debes ver 10 clases en TRAIN si tu estructura es tipo/severidad + sin_grietas
#   ['compresion_vertical_Grave', 'compresion_vertical_Leve', 'compresion_vertical_Moderada',
#    'friccion_cortante_escalonada_Grave', 'friccion_cortante_escalonada_Leve', 'friccion_cortante_escalonada_Moderada',
#    'tension_diagonal_inclinadas_Grave','tension_diagonal_inclinadas_Leve','tension_diagonal_inclinadas_Moderada',
#    'sin_grietas']

# Guardamos la lista de clases detectadas para la etapa multiclase (sin 'sin_grietas'):
multi_classes = [c for c in train_classes if _norm(c).lower() != "sin_grietas"]
multi_classes = sorted(multi_classes)  # orden estable
all_classes   = sorted(train_classes)  # incluye 'sin_grietas'

MODEL_DIR.mkdir(parents=True, exist_ok=True)
(Path(MODEL_DIR) / "classes_all.json").write_text(
    json.dumps({"class_names": all_classes}, ensure_ascii=False, indent=2),
    encoding="utf-8"
)
(Path(MODEL_DIR) / "classes_multi.json").write_text(
    json.dumps({"class_names": multi_classes}, ensure_ascii=False, indent=2),
    encoding="utf-8"
)
print("\nGuardado classes_all.json (10 esperadas) y classes_multi.json (9 esperadas, sin 'sin_grietas').")

```

**ConstrucciÃ³n de datasets (binario y multiclase) con tf.data**

```python
# === Celda 4 (completa): preparar splits binario + multiclase y guardar JSON de clases ===
from pathlib import Path
from collections import Counter, defaultdict
import numpy as np, glob, json, os

# --- Rutas (si ya las definiste antes, se respetan; si no, se crean por defecto) ---
TRAIN_DIR = globals().get("TRAIN_DIR", Path(r"C:\Users\User\mansory\train"))
VAL_DIR   = globals().get("VAL_DIR",   Path(r"C:\Users\User\mansory\test"))
MODEL_DIR = globals().get("MODEL_DIR", Path(r"C:\Users\User\mansory\modelos"))
MODEL_DIR.mkdir(parents=True, exist_ok=True)

IMG_EXTS = (".jpg",".jpeg",".png",".bmp",".webp",".JPG",".JPEG",".PNG")

def norm(s: str) -> str:
    """Normaliza nombres para comparar (sin guiones/guiones bajos, minÃºsculas)."""
    return " ".join(s.lower().replace("_"," ").replace("-"," ").split())

# Aliases posibles para la carpeta NEGATIVA (sin grietas)
NEG_ALIASES = {
    "sin grietas","sin fisuras","no grietas","no fisuras",
    "healthy","sano","intacto","no cracks","no crack"
}

def list_top_classes(root: Path):
    return sorted([p.name for p in root.iterdir() if p.is_dir()])

def rec_images(d: Path):
    """Obtiene imÃ¡genes recursivamente bajo un directorio."""
    files = []
    for ext in IMG_EXTS:
        files += glob.glob(str(d / "**" / f"*{ext}"), recursive=True)
    return sorted(files)

print("RaÃ­z TRAIN:", TRAIN_DIR)
print("RaÃ­z VAL  :", VAL_DIR)

# --- Comprobar clases top-level (primer nivel) ---
top_train = list_top_classes(TRAIN_DIR)
top_val   = list_top_classes(VAL_DIR)
print("TRAIN â€” clases top:", top_train)
print("VAL   â€” clases top:", top_val)
assert top_train == top_val, "Las clases top (carpetas de primer nivel) difieren entre train y test."

# --- Localizar carpeta negativa (sin grietas) por alias ---
neg_candidates = [c for c in top_train if norm(c) in NEG_ALIASES]
if not neg_candidates:
    raise AssertionError("No encuentro la clase negativa (por ejemplo 'sin grietas'). "
                         "Renombra tu carpeta negativa o agrega su alias en NEG_ALIASES.")
NEG_NAME = neg_candidates[0]                # nombre tal cual en disco (p.ej. 'sin grietas')
NEG_SAVE = NEG_NAME.replace(" ", "_")       # versiÃ³n estandarizada para JSON (p.ej. 'sin_grietas')
print(f"âœ… Negativa detectada: '{NEG_NAME}'")

# ======================
# 1) SPLIT BINARIO
# ======================
# 0 = sin grietas, 1 = con grieta
CAP_SIN_TRAIN = globals().get("CAP_SIN_TRAIN", 3000)  # recorte de negativos en train
CAP_SIN_VAL   = globals().get("CAP_SIN_VAL",   1500)  # recorte de negativos en val

train_files, train_labels = [], []
val_files,   val_labels   = [], []

# negativos (recursivo por si hay subcarpetas dentro de la negativa)
neg_train = rec_images(TRAIN_DIR / NEG_NAME)[:CAP_SIN_TRAIN]
neg_val   = rec_images(VAL_DIR   / NEG_NAME)[:CAP_SIN_VAL]
train_files.extend(neg_train); train_labels.extend([0]*len(neg_train))
val_files.extend(neg_val);     val_labels.extend([0]*len(neg_val))

# positivos: todas las carpetas excepto la negativa (recursivo)
pos_tops = [c for c in top_train if c != NEG_NAME]
for c in pos_tops:
    ftr = rec_images(TRAIN_DIR / c)
    fva = rec_images(VAL_DIR   / c)
    train_files.extend(ftr); train_labels.extend([1]*len(ftr))
    val_files.extend(fva);   val_labels.extend([1]*len(fva))

y_train_bin = np.array(train_labels, dtype=np.int32)
y_val_bin   = np.array(val_labels,   dtype=np.int32)

print(f"\nBINARIO â†’ train={len(y_train_bin)}  val={len(y_val_bin)}")
print("BIN train (0/1):", Counter(y_train_bin))
print("BIN val   (0/1):", Counter(y_val_bin))

# ======================
# 2) SPLIT MULTICLASE (9 clases: tipo_severidad; se excluye la carpeta NEGATIVA)
# ======================
def severities_under(root: Path, top: str):
    """Devuelve subcarpetas inmediatas (severidades) que contengan al menos 1 imagen."""
    sev = []
    topdir = root / top
    if not topdir.exists(): 
        return sev
    for sub in sorted([p.name for p in topdir.iterdir() if p.is_dir()]):
        if len(rec_images(topdir / sub)) > 0:
            sev.append(sub)
    return sev

# Unir severidades presentes en train y/o val (por cada tipo positivo)
type_sev = defaultdict(set)
for t in pos_tops:
    for s in severities_under(TRAIN_DIR, t):
        type_sev[t].add(s)
    for s in severities_under(VAL_DIR, t):
        type_sev[t].add(s)

# Construir lista de clases multiclase en orden estable
class_names_multi = []
for t in sorted(type_sev.keys()):
    for s in sorted(type_sev[t]):
        class_names_multi.append(f"{t}_{s}".replace(" ", "_"))

assert len(class_names_multi) > 0, (
    "No se detectaron clases multiclase (tipo+severidad). "
    "AsegÃºrate de tener subcarpetas (Grave/Leve/Moderada) con imÃ¡genes."
)

# mapa nombre â†’ Ã­ndice
name2id_multi = {n:i for i,n in enumerate(class_names_multi)}

# Recolectar archivos/labels multiclase (recursivo en tipo/severidad)
train_files2, y_train_multi = [], []
val_files2,   y_val_multi   = [], []

for t in sorted(type_sev.keys()):
    for s in sorted(type_sev[t]):
        cname = f"{t}_{s}".replace(" ", "_")
        idx = name2id_multi[cname]
        ftr = rec_images(TRAIN_DIR / t / s)
        fva = rec_images(VAL_DIR   / t / s)
        train_files2.extend(ftr); y_train_multi.extend([idx]*len(ftr))
        val_files2.extend(fva);   y_val_multi.extend([idx]*len(fva))

y_train_multi = np.array(y_train_multi, dtype=np.int32)
y_val_multi   = np.array(y_val_multi,   dtype=np.int32)

print("\nMULTICLASE â†’ clases (descubiertas):", class_names_multi)
print(f"MULTI â†’ train={len(y_train_multi)}  val={len(y_val_multi)}")
print("MULTI train por clase:", Counter(y_train_multi))
print("MULTI val   por clase:", Counter(y_val_multi))

# ======================
# 3) Guardar JSON de clases (estandarizando la negativa a 'sin_grietas')
# ======================
(MODEL_DIR / "classes_all.json").write_text(
    json.dumps({"class_names": [NEG_SAVE] + class_names_multi}, ensure_ascii=False, indent=2),
    encoding="utf-8",
)
(MODEL_DIR / "classes_multi.json").write_text(
    json.dumps({"class_names": class_names_multi}, ensure_ascii=False, indent=2),
    encoding="utf-8",
)

print("\nGuardado classes_all.json y classes_multi.json en", MODEL_DIR)

# --- Variables generadas por esta celda (para las siguientes) ---
#   NEG_NAME, NEG_SAVE
#   train_files, val_files, y_train_bin, y_val_bin
#   train_files2, val_files2, y_train_multi, y_val_multi
#   class_names_multi, name2id_multi


```

**Datasets y construcciÃ³n de modelos**

```python
# Celda 5: datasets y modelos

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
from pathlib import Path
import json, math
AUTOTUNE = tf.data.AUTOTUNE

# --- HiperparÃ¡metros comunes ---
IMG_SIZE   = (160, 160)
BATCH_SIZE = 32

# â€”â€”â€” util: lectura y preproceso
def _load_img(path):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.convert_image_dtype(img, tf.float32)       # [0,1]
    img = tf.image.resize(img, IMG_SIZE, antialias=True)
    img = (img - 0.5) * 2.0                                   # [-1,1] para MobileNetV2
    return img

def make_ds(files, labels=None, training=False):
    ds = tf.data.Dataset.from_tensor_slices(files)
    ds = ds.map(lambda p: _load_img(p), num_parallel_calls=AUTOTUNE)
    if training:
        aug = keras.Sequential([
            layers.RandomFlip("horizontal"),
            layers.RandomRotation(0.05),
            layers.RandomContrast(0.15),
        ])
        ds = ds.map(lambda x: aug(x, training=True), num_parallel_calls=AUTOTUNE)

    if labels is not None:
        lab = tf.data.Dataset.from_tensor_slices(labels)
        ds = tf.data.Dataset.zip((ds, lab))

    if training:
        ds = ds.shuffle(buffer_size=min(len(files), 2000), reshuffle_each_iteration=True)
    ds = ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)
    return ds

# === Datasets de esta sesiÃ³n (generados por Celda 4) ===
# BINARIO
train_ds_bin = make_ds(train_files, y_train_bin, training=True)
val_ds_bin   = make_ds(val_files,   y_val_bin,   training=False)

# MULTICLASE (solo imÃ¡genes con grietas)
train_ds_multi = make_ds(train_files2, y_train_multi, training=True)
val_ds_multi   = make_ds(val_files2,   y_val_multi,   training=False)

# === ConstrucciÃ³n de modelos ===
def build_backbone():
    inp  = layers.Input((*IMG_SIZE, 3))
    x    = inp  # ya normalizamos a [-1,1]
    base = keras.applications.MobileNetV2(
        include_top=False, weights="imagenet", input_shape=(*IMG_SIZE,3)
    )
    base.trainable = False
    x = base(x, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.25)(x)
    return inp, x, base

def build_binary_model():
    inp, feat, base = build_backbone()
    out = layers.Dense(1, activation="sigmoid", name="bin_out")(feat)
    model = keras.Model(inp, out, name="bin_mansory")
    model.compile(
        optimizer=keras.optimizers.Adam(1e-4),
        loss="binary_crossentropy",
        metrics=["accuracy", keras.metrics.AUC(name="auc")]
    )
    return model, base

def build_multi_model(num_classes: int):
    inp, feat, base = build_backbone()
    out = layers.Dense(num_classes, activation="softmax", name="multi_out")(feat)
    model = keras.Model(inp, out, name="multi_mansory")
    model.compile(
        optimizer=keras.optimizers.Adam(1e-5),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy", keras.metrics.AUC(name="auc", multi_label=False, num_labels=1)]
    )
    return model, base

# Rutas de guardado
MODEL_DIR   = Path(MODEL_DIR)
BIN_MODEL   = MODEL_DIR / "bin_sin_con.keras"
MULTI_MODEL = MODEL_DIR / "multi_grietas_9.keras"

print("IMG_SIZE:", IMG_SIZE, "| BATCH_SIZE:", BATCH_SIZE)
print("BIN_MODEL :", BIN_MODEL)
print("MULTI_MODEL:", MULTI_MODEL)

```

**Entrenamiento binario (sin/con grieta)**

```python
# Celda 6: entrenamiento BINARIO (0=sin, 1=con grieta)

from collections import Counter
import numpy as np

# Pesos por clase (inverso de frecuencia)
classes, counts = np.unique(y_train_bin, return_counts=True)
total = counts.sum()
cw_bin = {int(c): float(total / (len(classes) * n)) for c, n in zip(classes, counts)}
print("class_weight (bin) usado :", cw_bin)

bin_model, bin_base = build_binary_model()

cbs = [
    keras.callbacks.EarlyStopping(patience=8, restore_best_weights=True, monitor="val_accuracy"),
    keras.callbacks.ReduceLROnPlateau(patience=3, factor=0.5, monitor="val_loss"),
    keras.callbacks.ModelCheckpoint(str(BIN_MODEL), monitor="val_accuracy", save_best_only=True)
]

hist_bin = bin_model.fit(
    train_ds_bin,
    validation_data=val_ds_bin,
    epochs=30,
    class_weight=cw_bin,
    callbacks=cbs,
    verbose=1
)
print("\nMejor val_acc (bin):", np.max(hist_bin.history["val_accuracy"]))

```

```python
# Celda 7A: modelo multiclase (9 clases) â€” sin AUC

from tensorflow import keras
from tensorflow.keras import layers

# AsegÃºrate de tener estas variables de celdas previas:
# - class_names_multi (lista con las 9 clases)
# - IMG_SIZE (H, W)
# - MULTI_MODEL (ruta .keras)
# - make_ds, train_files2, val_files2, y_train_multi, y_val_multi
# - cw_multi (class_weight multiclase)  -> si no existe, lo calculamos abajo

H, W = IMG_SIZE
C = len(class_names_multi)
print("Clases multiclase (C):", C, "|", class_names_multi)

def build_multi_model(num_classes: int):
    inp  = keras.Input(shape=(H, W, 3))
    base = keras.applications.MobileNetV2(
    include_top=False, weights="imagenet", input_shape=(H, W, 3), input_tensor=inp
)

    
    base.trainable = False

    x = layers.GlobalAveragePooling2D()(base.output)
    x = layers.Dropout(0.35)(x)
    x = layers.Dense(256, activation="relu")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.35)(x)
    out = layers.Dense(num_classes, activation="softmax", name="multi_out")(x)

    model = keras.Model(inp, out, name="multi_mansory")

    # ðŸ‘‡ MÃ©tricas correctas para multiclase con etiquetas "sparse"
    model.compile(
        optimizer=keras.optimizers.Adam(1e-5),
        loss="sparse_categorical_crossentropy",
        metrics=[
            "accuracy",
            keras.metrics.SparseTopKCategoricalAccuracy(k=3, name="top3"),
        ],
    )
    return model, base

# Datasets (si no existen aÃºn)
try:
    train_ds_multi
    val_ds_multi
except NameError:
    train_ds_multi = make_ds(train_files2, y_train_multi, shuffle=True)
    val_ds_multi   = make_ds(val_files2,   y_val_multi,   shuffle=False)

# class_weight multiclase (si no lo calculaste antes)
try:
    cw_multi
except NameError:
    import numpy as np
    from collections import Counter
    cnt = Counter(y_train_multi)
    tot = len(y_train_multi)
    cls = [i for i in range(C)]
    cw_multi = {int(c): float(min(50.0, tot / (C * cnt.get(c, 1)))) for c in cls}
    print("class_weight (multi) usado:", cw_multi)

multi_model, multi_base = build_multi_model(C)


```

**entrenar multiclase (con callbacks)**

```python
# === Celda 7B-0 (shim de compatibilidad para make_ds) ===
import inspect

def mkds(files, labels, batch_size, img_size, augment):
    """
    Envuelve make_ds con compatibilidad de firmas:
    - Si make_ds(..., batch=) -> usamos batch=
    - Si make_ds(..., batch_size=) -> usamos batch_size=
    - Si no acepta kwargs -> usamos posicional
    """
    params = inspect.signature(make_ds).parameters
    if "batch" in params:
        return make_ds(files, labels, batch=batch_size, img_size=img_size, augment=augment)
    if "batch_size" in params:
        return make_ds(files, labels, batch_size=batch_size, img_size=img_size, augment=augment)
    # Fallback posicional
    try:
        return make_ds(files, labels, batch_size, img_size, augment)
    except TypeError:
        # Ãšltimo recurso por si tu make_ds ignora augment o img_size
        try:
            return make_ds(files, labels, batch_size, img_size)
        except TypeError:
            return make_ds(files, labels)

```

```python
# === Celda 7B-FIX (nueva): constructor de datasets para multiclase ===
import tensorflow as tf
AUTOTUNE = tf.data.AUTOTUNE

def _load_and_preprocess(path, label, img_size):
    H, W = img_size
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3, expand_animations=False)
    img = tf.image.convert_image_dtype(img, tf.float32)   # [0,1]
    img = tf.image.resize(img, [H, W])
    img = (img - 0.5) * 2.0                               # [-1,1] igual que binario
    return img, label

def _augment(img, label):
    img = tf.image.random_flip_left_right(img)
    img = tf.image.random_brightness(img, 0.10)
    img = tf.image.random_contrast(img, 0.90, 1.10)
    img = tf.image.random_saturation(img, 0.90, 1.10)
    return img, label

def build_ds(files, labels, batch_size, img_size, augment=False, shuffle=True):
    ds = tf.data.Dataset.from_tensor_slices((files, labels))
    if shuffle:
        ds = ds.shuffle(min(len(files), 10000), reshuffle_each_iteration=True)
    ds = ds.map(lambda f, l: _load_and_preprocess(f, l, img_size),
                num_parallel_calls=AUTOTUNE)
    if augment:
        ds = ds.map(_augment, num_parallel_calls=AUTOTUNE)
    ds = ds.batch(batch_size)
    ds = ds.prefetch(AUTOTUNE)
    return ds
```

```python
# === Celda 7B (LISTA Y ROBUSTA) â€” Multiclase SOLO la cabeza ===
import tensorflow as tf
from tensorflow import keras
from keras import layers
import numpy as np
from pathlib import Path

# ---- Comprobaciones de variables previas (de Celdas 4 y 5) ----
assert 'train_files2' in globals() and 'val_files2' in globals(), "Faltan train_files2/val_files2 (Celda 4)."
assert 'y_train_multi' in globals() and 'y_val_multi' in globals(), "Faltan y_train_multi/y_val_multi (Celda 4)."
assert 'class_names_multi' in globals(), "Falta class_names_multi (Celda 4)."
assert 'IMG_SIZE' in globals() and 'BATCH_SIZE' in globals(), "Faltan IMG_SIZE/BATCH_SIZE (Celda 5)."
assert 'MODEL_DIR' in globals(), "Falta MODEL_DIR (Celda 5)."

# ---- Etiquetas a int32 ----
y_train_multi = np.asarray(y_train_multi, dtype=np.int32)
y_val_multi   = np.asarray(y_val_multi,   dtype=np.int32)

C = len(class_names_multi)
print("Clases multiclase (C):", C, "|", class_names_multi)

# ============================================
# Datasets propios (normalizaciÃ³n = [-1, 1])
# ============================================
AUTOTUNE = tf.data.AUTOTUNE

def _load_and_preprocess(path, label, img_size):
    H, W = img_size
    img  = tf.io.read_file(path)
    img  = tf.image.decode_image(img, channels=3, expand_animations=False)
    img  = tf.image.convert_image_dtype(img, tf.float32)   # [0,1]
    img  = tf.image.resize(img, [H, W])
    img  = (img - 0.5) * 2.0                                # [-1,1] (consistente con binario)
    return img, label

def _augment(img, label):
    img = tf.image.random_flip_left_right(img)
    img = tf.image.random_brightness(img, 0.10)
    img = tf.image.random_contrast(img, 0.90, 1.10)
    img = tf.image.random_saturation(img, 0.90, 1.10)
    return img, label

def build_ds(files, labels, batch_size, img_size, augment=False, shuffle=True):
    ds = tf.data.Dataset.from_tensor_slices((files, labels))
    if shuffle:
        ds = ds.shuffle(min(len(files), 10000), reshuffle_each_iteration=True)
    ds = ds.map(lambda f, l: _load_and_preprocess(f, l, img_size), num_parallel_calls=AUTOTUNE)
    if augment:
        ds = ds.map(_augment, num_parallel_calls=AUTOTUNE)
    ds = ds.batch(batch_size).prefetch(AUTOTUNE)
    return ds

train_ds_multi = build_ds(train_files2, y_train_multi, BATCH_SIZE, IMG_SIZE, augment=True,  shuffle=True)
val_ds_multi   = build_ds(val_files2,   y_val_multi,   BATCH_SIZE, IMG_SIZE, augment=False, shuffle=False)

xb, yb = next(iter(train_ds_multi))
print("Batch ejemplo:", xb.shape, yb.shape, yb.dtype)

# =========================
# Pesos por clase (capados)
# =========================
tot = len(y_train_multi)
cls, cnt = np.unique(y_train_multi, return_counts=True)
cw_multi = {int(c): float(min(30.0, tot / (C * n))) for c, n in zip(cls, cnt)}
print("class_weight (multi) usado:", cw_multi)

# =========================
# Modelo MULTICLASE (cabeza)
# =========================
H, W = IMG_SIZE
inp  = layers.Input(shape=(H, W, 3))
base = keras.applications.MobileNetV2(
    include_top=False,
    weights="imagenet",
    input_shape=(H, W, 3),
    input_tensor=inp
)
base.trainable = False  # Entrenamos SOLO la cabeza

x   = base.output
x   = layers.GlobalAveragePooling2D()(x)
x   = layers.Dropout(0.35)(x)
out = layers.Dense(C, activation="softmax", name="multi_out")(x)
multi_model = keras.Model(inp, out, name="multi_mansory_multi")

# ---- CompilaciÃ³n: intentar label_smoothing; si no existe, fallback sin smoothing ----
try:
    loss_obj = keras.losses.SparseCategoricalCrossentropy(label_smoothing=0.1)
    print("Usando SparseCategoricalCrossentropy con label_smoothing=0.1")
except TypeError:
    loss_obj = keras.losses.SparseCategoricalCrossentropy()
    print("Tu tf.keras no soporta label_smoothing en SparseCategoricalCrossentropy. Uso sin smoothing.")

multi_model.compile(
    optimizer=keras.optimizers.Adam(1e-3),
    loss=loss_obj,
    metrics=[
        "accuracy",
        keras.metrics.SparseTopKCategoricalAccuracy(k=3, name="top3"),
    ],
)

# =========================
# Callbacks y entrenamiento
# =========================
MULTI_MODEL = MODEL_DIR / "multi_grietas_9.keras"
cbs2 = [
    keras.callbacks.EarlyStopping(patience=6, restore_best_weights=True, monitor="val_accuracy"),
    keras.callbacks.ReduceLROnPlateau(patience=2, factor=0.5, monitor="val_loss"),
    keras.callbacks.ModelCheckpoint(str(MULTI_MODEL), monitor="val_accuracy", save_best_only=True),
]

hist_multi = multi_model.fit(
    train_ds_multi,
    validation_data=val_ds_multi,
    epochs=20,
    class_weight=cw_multi,
    verbose=1,
    callbacks=cbs2,
)

print("\nMejor val_acc (multi, cabeza):", max(hist_multi.history["val_accuracy"]))
print("Modelo multiclase guardado en:", MULTI_MODEL)

```

**Fine-tuning parcial (cola del backbone + BN congeladas)**

```python
# === Celda 7C: Fine-tuning parcial (sin backbone.layers) ===
import tensorflow as tf
from tensorflow import keras
from keras import layers
from pathlib import Path

assert 'multi_model' in globals(), "Primero ejecuta la Celda 7B (entrenar la cabeza)."
assert 'train_ds_multi' in globals() and 'val_ds_multi' in globals(), "Faltan datasets multiclase (Celda 7B)."
assert 'cw_multi' in globals(), "Faltan class_weight multiclase (Celda 7B)."

# 1) Congelar TODO inicialmente
for l in multi_model.layers:
    l.trainable = False

# 2) Descongelar SOLO las Ãºltimas N capas convolucionales (cola)
N = 60  # prueba 60; si no mejora, intenta 80
conv_types = (layers.Conv2D, layers.DepthwiseConv2D, getattr(layers, "SeparableConv2D", tuple()))
bn_type = layers.BatchNormalization

unfrozen_convs = 0
for l in reversed(multi_model.layers):
    # Mantener BN siempre congeladas (evita drift con lotes pequeÃ±os)
    if isinstance(l, bn_type):
        l.trainable = False
        continue
    # Descongelar Ãºltimas N capas de tipo conv/depthwise/separable
    if isinstance(l, conv_types):
        if unfrozen_convs < N:
            l.trainable = True
            unfrozen_convs += 1
        else:
            l.trainable = False

# 3) Asegurar que la cabeza (Dense final) quede entrenable
for l in multi_model.layers:
    if isinstance(l, layers.Dense):
        l.trainable = True

# 4) Contar cuÃ¡ntas capas quedaron entrenables
trainable_cnt = sum(1 for l in multi_model.layers if l.trainable)
total_cnt     = len(multi_model.layers)
print(f"Capas entrenables: {trainable_cnt}/{total_cnt}  | convs desbloqueadas: {unfrozen_convs}")

# 5) Recompilar con LR bajo
try:
    loss_obj = keras.losses.SparseCategoricalCrossentropy(label_smoothing=0.1)
except TypeError:
    loss_obj = keras.losses.SparseCategoricalCrossentropy()

multi_model.compile(
    optimizer=keras.optimizers.Adam(1e-5),
    loss=loss_obj,
    metrics=[
        "accuracy",
        keras.metrics.SparseTopKCategoricalAccuracy(k=3, name="top3"),
    ],
)

# 6) Entrenar fine-tuning
MULTI_MODEL = MODEL_DIR / "multi_grietas_9.keras"
cbs_ft = [
    keras.callbacks.EarlyStopping(patience=6, restore_best_weights=True, monitor="val_accuracy"),
    keras.callbacks.ReduceLROnPlateau(patience=2, factor=0.5, monitor="val_loss"),
    keras.callbacks.ModelCheckpoint(str(MULTI_MODEL), monitor="val_accuracy", save_best_only=True),
]

hist_ft = multi_model.fit(
    train_ds_multi,
    validation_data=val_ds_multi,
    epochs=15,              # baja si tu CPU estÃ¡ al 100%
    class_weight=cw_multi,  # mantiene pesos para desbalance
    verbose=1,
    callbacks=cbs_ft,
)

print("\n(FT) Mejor val_acc (multi):", max(hist_ft.history["val_accuracy"]))
print("Modelo multiclase guardado en:", MULTI_MODEL)

```

**Inferencia two-stage (carga modelos + umbral + helpers)**

```python
# === Celda 9: inferencia two-stage con umbral guardado ===
from pathlib import Path
import json, numpy as np
from tensorflow import keras
from keras.utils import load_img, img_to_array

MODEL_DIR   = Path("modelos")
BIN_PATH    = MODEL_DIR / "bin_sin_con.keras"
MULTI_PATH  = MODEL_DIR / "multi_grietas_9.keras"
TH_PATH     = MODEL_DIR / "thresholds.json"
CL_ALL      = MODEL_DIR / "classes_all.json"     # [ 'sin_grietas', <9 clases ...> ]
CL_MULTI    = MODEL_DIR / "classes_multi.json"   # [ <9 clases ...> ]

assert BIN_PATH.exists(),   f"No existe {BIN_PATH}"
assert MULTI_PATH.exists(), f"No existe {MULTI_PATH}"

# 1) Cargar modelos (sin compilar para evitar problemas de 'loss')
bin_model   = keras.models.load_model(str(BIN_PATH), compile=False)
multi_model = keras.models.load_model(str(MULTI_PATH), compile=False)

# 2) Cargar umbral (fallback si no existe)
if TH_PATH.exists():
    BIN_THRESHOLD = float(json.loads(TH_PATH.read_text(encoding="utf-8"))["bin_threshold"])
else:
    BIN_THRESHOLD = 0.83  # fallback
print(f"Umbral binario usado: {BIN_THRESHOLD:.3f}")

# 3) Cargar nombres de clases
class_names_all   = json.loads(CL_ALL.read_text(encoding="utf-8"))["class_names"]
class_names_multi = json.loads(CL_MULTI.read_text(encoding="utf-8"))["class_names"]

# 4) Construir mapeo por tipo (agregaciÃ³n tipo â†’ severidades)
#    (toma todo lo que estÃ¡ antes del Ãºltimo "_")
def split_type_sev(name: str):
    if "_" not in name: 
        return name, ""
    t, s = name.rsplit("_", 1)
    return t, s

types = {}             # tipo -> lista de (idx, nombre_clase_multi, severidad)
for idx, cname in enumerate(class_names_multi):
    t, s = split_type_sev(cname)
    types.setdefault(t, []).append((idx, cname, s))

print("Tipos detectados:", list(types.keys()))

# 5) Preprocesado acorde al input del modelo
def _prep_for(model, pil_img):
    H, W = model.input_shape[1], model.input_shape[2]
    arr = img_to_array(load_img(pil_img, target_size=(H, W)))
    arr = arr.astype("float32") / 255.0   # normalizaciÃ³n usada en entrenamiento
    arr = np.expand_dims(arr, 0)
    return arr

# 6) PredicciÃ³n two-stage con decisiÃ³n por tipoâ†’severidad
def predict_two_stage_type_first(img_path: str):
    # Paso 1: binario (Â¿hay grieta?)
    x_bin = _prep_for(bin_model, img_path)
    p_con = float(bin_model.predict(x_bin, verbose=0)[0][0])  # prob. "con grieta" (clase=1)
    if p_con < BIN_THRESHOLD:
        return {
            "tipo": "sin_grietas",
            "severidad": "â€”",
            "conf_tipo": 1.0 - p_con,
            "conf_severidad": 1.0,
            "detalle": {"p_con_grieta": p_con}
        }

    # Paso 2: multiclase (9 clases)
    x_mul = _prep_for(multi_model, img_path)
    probs = multi_model.predict(x_mul, verbose=0)[0]  # vector softmax (len = 9)

    # Agregar por tipo
    type_scores = {}
    for t, lst in types.items():
        type_scores[t] = float(sum(probs[i] for i, _, _ in lst))

    # Elegir mejor tipo
    best_type = max(type_scores, key=type_scores.get)
    conf_type = type_scores[best_type]

    # Elegir severidad dentro del mejor tipo
    cand = [(i, cname, sev, probs[i]) for (i, cname, sev) in types[best_type]]
    cand.sort(key=lambda x: x[3], reverse=True)
    best_idx, best_cname, best_sev, best_p = cand[0]

    # Formatear bonitos (cambiar "_" por espacio)
    fmt_type = best_type.replace("_", " ")
    fmt_sev  = best_sev.replace("_", " ")

    return {
        "tipo": fmt_type,
        "severidad": fmt_sev,
        "conf_tipo": conf_type,
        "conf_severidad": float(best_p),
        "detalle": {
            "p_con_grieta": p_con,
            "top3_clases": sorted(
                [(class_names_multi[i], float(p)) for i, p in enumerate(probs)],
                key=lambda x: x[1], reverse=True
            )[:3]
        }
    }
```

```python
# Celda 9B â€” detector por mosaicos + multi-escala (sin OpenCV)
from tensorflow import keras
from keras.utils import load_img, img_to_array
import numpy as np, math, json
from pathlib import Path

# === Cargar modelos y clases (si no estÃ¡n en memoria) ===
MODEL_DIR = Path("modelos")
BIN_MODEL   = MODEL_DIR / "bin_sin_con.keras"
MULTI_MODEL = MODEL_DIR / "multi_grietas_9.keras"

bin_model   = bin_model   if "bin_model"   in globals() else keras.models.load_model(str(BIN_MODEL))
multi_model = multi_model if "multi_model" in globals() else keras.models.load_model(str(MULTI_MODEL))

with open(MODEL_DIR/"classes_multi.json","r",encoding="utf-8") as f:
    class_names_multi = json.load(f)["class_names"]

# Umbral (puedes sobreescribirlo con la celda 9A-bis)
if "BIN_THRESHOLD" not in globals():
    BIN_THRESHOLD = 0.5

def _prep_np(pil_img, H, W):
    x = img_to_array(pil_img.resize((W, H))) / 255.0
    return np.expand_dims(x.astype("float32"), 0)

def _predict_bin_np(arr):
    p = float(bin_model.predict(arr, verbose=0)[0][0])  # prob "con grieta"
    return p

def _predict_multi_np(arr):
    p = multi_model.predict(arr, verbose=0)[0]  # softmax
    idx = int(np.argmax(p))
    return idx, float(p[idx]), p

def tile_scan(path, grid=(3,5), overlap=0.3, scales=(1.0, 1.25, 1.6), k_top=3):
    """
    Devuelve: dict con mejor prob binaria, bbox del mejor mosaico, y votaciÃ³n multiclase.
    """
    from PIL import Image
    im0 = Image.open(path).convert("RGB")
    Hm, Wm = bin_model.input_shape[1], bin_model.input_shape[2]  # 160x160
    best = {"p_con": -1.0, "tile": None, "scale": None, "prob_vec": None}

    all_multi = []

    for sc in scales:
        W = int(im0.width * sc)
        H = int(im0.height * sc)
        im = im0.resize((W, H))

        gy, gx = grid
        # tamaÃ±o base del mosaico (con solape)
        tw = int(W / (gx - (gx-1)*overlap))
        th = int(H / (gy - (gy-1)*overlap))
        sx = int(tw * (1-overlap))
        sy = int(th * (1-overlap))

        tiles = []
        coords = []
        for y in range(0, max(1, H - th + 1), max(1, sy)):
            for x in range(0, max(1, W - tw + 1), max(1, sx)):
                crop = im.crop((x, y, x+tw, y+th))
                arr = _prep_np(crop, Hm, Wm)
                p = _predict_bin_np(arr)
                tiles.append((p, crop, arr))
                coords.append((x, y, x+tw, y+th, sc))

        # ordenar por probabilidad y quedarnos con los mejores k para multiclase
        tiles_sorted = sorted(zip(tiles, coords), key=lambda z: z[0][0], reverse=True)
        for (p, crop, arr), c in tiles_sorted[:k_top]:
            if p > best["p_con"]:
                best.update({"p_con": p, "tile": c, "scale": sc})
            # multiclase solo si parece haber grieta
            if p >= BIN_THRESHOLD:
                idx, conf, prob = _predict_multi_np(arr)
                all_multi.append((idx, conf, prob))

    # decisiÃ³n binaria agregada (max)
    is_crack = best["p_con"] >= BIN_THRESHOLD

    # voto multiclase (promedio de probas de los tiles con grieta)
    multi_idx = None; multi_conf = 0.0; multi_name = None
    if all_multi:
        probs = np.stack([p for _,_,p in all_multi], axis=0)  # [K, C]
        meanp = probs.mean(axis=0)
        multi_idx = int(np.argmax(meanp))
        multi_conf = float(meanp[multi_idx])
        multi_name = class_names_multi[multi_idx]

    return {
        "is_crack": is_crack,
        "p_con_max": float(best["p_con"]),
        "best_tile": best["tile"],    # (x1,y1,x2,y2,scale) sobre la imagen escalada
        "type_idx": multi_idx,
        "type_name": multi_name,
        "type_conf": multi_conf,
    }

def predict_two_stage_tiled(path):
    r = tile_scan(path, grid=(3,5), overlap=0.35, scales=(1.0,1.3,1.6), k_top=5)
    if not r["is_crack"]:
        return {
            "tipo": "sin_grietas",
            "severidad": "â€”",
            "conf_tipo": 1.0 - r["p_con_max"],
            "extra": r,
        }
    # mapear nombre multiclase a (tipo, severidad)
    name = r["type_name"]
    tipo, severidad = name.split("_", 1) if name and "_" in name else (name, "â€”")
    return {
        "tipo": tipo,
        "severidad": severidad,
        "conf_tipo": r["type_conf"],
        "extra": r,
    }

# === Prueba con tus dos imÃ¡genes ===
TEST_IMGS = [
    r"C:\Users\User\mansory\grieta-muro.jpg",
    # aÃ±ade mÃ¡s rutas si quieres
]
for p in TEST_IMGS:
    out = predict_two_stage_tiled(p)
    print("\nArchivo:", p)
    print("  Resultado â†’", out["tipo"], "|", out["severidad"])
    print(f"  p_con_max: {out['extra']['p_con_max']:.3f} (umbral={BIN_THRESHOLD:.2f})  |  conf tipo: {out['conf_tipo']:.3f}")
```
